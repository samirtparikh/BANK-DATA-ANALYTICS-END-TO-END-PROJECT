```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

numerical_cols = [
    'Estimated Income', 
    'Superannuation Savings', 
    'Credit Card Balance', 
    'Bank Loans', 
    'Bank Deposits', 
    'Checking Accounts', 
    'Saving Accounts', 
    'Foreign Currency Account', 
    'Business Lending'
]

# Generate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Format the correlation matrix values to two decimal places as strings
formatted_matrix = correlation_matrix.applymap(lambda x: f"{x:.2f}")

# Create a figure
fig, ax = plt.subplots(figsize=(12, 12))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f", ax=ax, cbar=False)
ax.set_title("Correlation Matrix")

# Create a table below the heatmap
table_ax = fig.add_subplot(212)  # Add a new subplot for the table
table = table_ax.table(cellText=formatted_matrix.values, 
                       colLabels=formatted_matrix.columns, 
                       rowLabels=formatted_matrix.index, 
                       cellLoc='center', 
                       loc='center')

# Adjust the table properties
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
table_ax.axis('off')  # Hide the axes for the table

# Adjust layout to make room for the table
plt.tight_layout()
plt.show()
```

The correlation table shows the relationships between various financial metrics. Here are some insights based on the correlations:

Insights from the Correlation Table
Strong Correlations:

Checking Accounts and Bank Deposits (0.84): There is a strong positive correlation, indicating that individuals with more funds in checking accounts tend to have higher bank deposits.
Bank Deposits and Saving Accounts (0.75): This suggests that people with higher bank deposits also tend to maintain substantial savings.
Moderate Correlations:

Estimated Income and Bank Loan (0.33): A moderate positive correlation suggests that higher estimated income may be associated with larger bank loans.
Credit Card Balance and Bank Deposits (0.38): Indicates that individuals with higher credit card balances may also have significant deposits, which could suggest a reliance on credit.

Weak Correlations:

Superannuation Savings and Business Lending (0.26): The weak correlation indicates limited relationship between these two metrics, suggesting that one does not significantly influence the other.
Saving Accounts and Foreign Currency Account (0.31): A weak correlation suggests that having savings does not strongly predict foreign currency holdings.

Negative Correlations:
There are no significant negative correlations in this dataset, which is somewhat typical in financial metrics, as most relationships tend to be positive.
Overall Trends
Financial Interdependence: The strong correlations among bank deposits, checking accounts, and savings accounts suggest a trend where individuals with higher liquidity in one area tend to have higher liquidity in others.
Income Influence: The moderate correlations with estimated income imply that income plays a role in determining borrowing behaviors, as seen with loans and credit card balances.
Conclusion
The correlation matrix highlights the interconnectedness of various banking metrics. Financial institutions can leverage this information to tailor products and services, such as promoting savings accounts to clients with significant checking balances. Understanding these correlations can also help individuals make informed financial decisions based on their existing banking behaviors.

The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts", and "Foreign Currency Account" including that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
